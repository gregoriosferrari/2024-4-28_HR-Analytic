# ABSA using Pretrained Model

## Create Instance

```python
from pyabsa import AspectTermExtraction as ATEPC
```

```python
aspect_extractor2 = ATEPC.AspectExtractor('multilingual2',
                                         auto_device=True,  # False means load model on CPU
                                         cal_perplexity=True,
                                         )
```

## Inference

```python
import os
import json
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HR_Analytic_From_gregoriosferrari2000/2024-4-24_preprocessedUlasan.csv')
```

```python
df.info()
df = df.dropna(subset=['good_translate'])
df.info()
df = df.copy()
```

```python
#@title Inference for split_sentence
import ast
import time

# Define column names
columns = ['Row_Num','sentence','IOB','tokens','aspect','position','sentiment','probs','confidence']
# Create an empty DataFrame with specified column names
df_sentiment = pd.DataFrame(columns=columns)

batch = 10

t1 = time.time()
for i in range(0, len(df), batch):
  # Get 10 rows of 'ulasan'
  df_batch = df[['Row_Num', 'split_sentence']].iloc[i:i+batch]

  # For split_sentence
  review_sentence_list = []
  row_num_list = []
  for df_batch_idx, df_batch_row in df_batch.iterrows():
    review_sentence_temp = ast.literal_eval(df_batch_row['split_sentence'])
    row_num_temp = df_batch_row['Row_Num']
    for review in review_sentence_temp:
      review_sentence_list.append(review)
      row_num_list.append(row_num_temp)
  print(row_num_list)

  # Predict sentiment
  absa_res = aspect_extractor2.predict(review_sentence_list,
                    save_result=True,
                    print_result=True,  # print the result
                    ignore_error=True,  # ignore the error when the model cannot predict the input
                    )
  json_data = absa_res

  # Iterate over the JSON data and add "row_num" key
  for i, item in enumerate(json_data):
      item['Row_Num'] = row_num_list[i]

  # Append data from JSON object as a row to the DataFrame
  df_sentiment = df_sentiment._append(json_data, ignore_index=True)

t2 = time.time()
```

```python
#@title Inference for good_translate
import ast
import time

# Define column names
columns = ['Row_Num','sentence','IOB','tokens','aspect','position','sentiment','probs','confidence']
# Create an empty DataFrame with specified column names
df_sentiment = pd.DataFrame(columns=columns)

batch = 10

t1 = time.time()
for i in range(0, len(df), batch):
  # Get 10 rows of 'ulasan'
  df_batch = df[['Row_Num', 'good_translate']].iloc[i:i+batch]

  # For split_sentence
  review_sentence_list = []
  row_num_list = []
  for df_batch_idx, df_batch_row in df_batch.iterrows():
    review_sentence_temp = df_batch_row['good_translate']
    row_num_temp = df_batch_row['Row_Num']

    review_sentence_list.append(review_sentence_temp)
    row_num_list.append(row_num_temp)
  print(row_num_list)

  # Predict sentiment
  absa_res = aspect_extractor2.predict(review_sentence_list,
                    save_result=True,
                    print_result=True,  # print the result
                    ignore_error=True,  # ignore the error when the model cannot predict the input
                    )
  json_data = absa_res

  # Iterate over the JSON data and add "row_num" key
  for i, item in enumerate(json_data):
      item['Row_Num'] = row_num_list[i]

  # Append data from JSON object as a row to the DataFrame
  df_sentiment = df_sentiment._append(json_data, ignore_index=True)

t2 = time.time()
```
